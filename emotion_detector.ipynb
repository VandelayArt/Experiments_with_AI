```python
import cv2
```

```python
from deepface import DeepFace
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('img/img-2.jpg')
```

```python
plt.imshow(img)
```

```python
predictions = DeepFace.analyze(img)
```

```python
predictions
```

```python
gender = predictions['gender']
age = predictions['age']
emotion = predictions['dominant_emotion']
race = predictions['dominant_race']

print(gender, age, emotion, race)
```

```python
facecascade = cv2.CascadeClassifier('data/haarcascade_frontalfaces_default.xml')
```

```python
from random import randrange

gs1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


# Detecting the face
faces = facecascade.detectMultiScale(gs1)

for (x, y, w, h) in faces:
    cv2.rectangle(img, (x,y), (x+w, y+w), (randrange(256), randrange(256), randrange(256)), 3)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX

cv2.putText(img,
                emotion,
                (0,50),
                font, 10,
                (randrange(256), randrange(256), randrange(256)),
                2, cv2.LINE_4)
```
